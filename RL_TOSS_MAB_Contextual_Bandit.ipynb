```python
# -*- coding: utf-8 -*-
"""
Contextual Bandit — LinUCB (No Oracle Line Version)

- Toss 광고 데이터에 맞춘 Contextual LinUCB 구현
- Reward 모델(SGDClassifier)을 이용해 P(click | context, arm) 을 만들고
  이를 "환경"의 진짜 CTR로 사용해 클릭을 샘플링하는 generative bandit.

[평가지표 출력]
  1) Cumulative CTR (그래프)
  2) Cumulative Regret vs best single arm (그래프)
  3) Estimated Q-value convergence: True CTR vs Q(a) (그래프)
  4) Action Selection Distribution: 팔 선택 count + 시간에 따른 누적 비율 (그래프)
  5) Context-conditioned CTR (예: day_of_week) (표)
  6) Policy Accuracy (정책 정확도, %)
  7) Model Loss (supervised reward model log-loss, arm별 + 평균)
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, Normalizer
from sklearn.feature_extraction import FeatureHasher
from sklearn.linear_model import SGDClassifier
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import log_loss

# tqdm은 진행상황 표시용 (선택)
try:
    from tqdm import tqdm
    TQDM_AVAILABLE = True
except ImportError:
    TQDM_AVAILABLE = False


# ---------------- CONFIG ----------------
PATH = "Toss_preprocessed_FF.csv"

CTX_COLS = [
    "gender",
    "age_group",
    "inventory_id",
    "day_of_week",
    "hour",
    "seq_last",
    "seq_len",
    "seq_entropy",
]

ARM_COL = "ad_contents"
REWARD_COL = "reward"

HASH_DIM = 128        # FeatureHasher 차원
TEST_SIZE_ENV = 0.30  # 전체 데이터 중 30%를 bandit 환경으로 사용
SEED = 42
ALPHA = 0.8           # LinUCB 탐색 상수
LOG_EVERY = 2000      # 몇 step마다 로그/시각화용 포인트 저장

PRED_CLIP = (1e-3, 1 - 1e-3)  # supervised 모델 확률 clipping
MIN_ARM_SAMPLES = 200         # supervised 모델을 학습하기 위한 최소 샘플 수

CONTEXT_SEG_COL = "day_of_week"   # Context-conditioned CTR 을 볼 기준 컬럼


# ---------------- SEED ----------------
np.random.seed(SEED)


# ---------------- LOAD DATA ----------------
df_raw = pd.read_csv(PATH)
df = df_raw.dropna(subset=CTX_COLS + [ARM_COL, REWARD_COL]).copy()
df[REWARD_COL] = df[REWARD_COL].astype(int).clip(0, 1)

print(f"[INFO] total rows={len(df):,}, unique arms={df[ARM_COL].nunique()}")


# ---------------- SPLIT: model vs env ----------------
model_df, env_df = train_test_split(
    df,
    test_size=TEST_SIZE_ENV,
    random_state=SEED,
    shuffle=True,
)

print(f"[SPLIT] model rows={len(model_df):,}, env rows={len(env_df):,}")

arms = np.sort(df[ARM_COL].unique())
K = len(arms)


# =====================================================
# 1. SUPERVISED REWARD MODEL (per arm)  + BCE LOSS
# =====================================================
models = {}
scalers = {}
arm_val_losses = []
arm_val_ns = []

for arm in arms:
    sub = model_df[model_df[ARM_COL] == arm]
    n_arm = len(sub)
    mean_ctr = sub[REWARD_COL].mean() if n_arm > 0 else 0.0
    print(f"[MODEL] arm={arm} n={n_arm:,}, mean_ctr={mean_ctr:.4f}")

    if n_arm < MIN_ARM_SAMPLES:
        # 데이터가 너무 적으면 상수 확률 모델로 대체
        p_const = float(np.clip(mean_ctr, *PRED_CLIP))
        models[arm] = ("const", p_const)
        continue

    # Numeric, Categorical 분리 (간단 버전)
    num_cols = ["seq_last", "seq_len", "seq_entropy"]
    cat_cols = ["gender", "age_group", "day_of_week", "hour"]

    X_num = sub[num_cols].values.astype(float)
    X_cat = sub[cat_cols].values.astype(float)  # one-hot 대신 그대로 사용
    X_all = np.hstack([X_num, X_cat])
    y_all = sub[REWARD_COL].values

    # train / valid split
    X_tr, X_val, y_tr, y_val = train_test_split(
        X_all, y_all, test_size=0.2, random_state=SEED
    )

    sc = StandardScaler().fit(X_tr)
    X_tr_std = sc.transform(X_tr)
    X_val_std = sc.transform(X_val)

    base = SGDClassifier(
        loss="log_loss",
        alpha=1e-4,
        max_iter=200,
        tol=1e-3,
        random_state=SEED,
    )
    # CalibratedClassifierCV으로 확률 보정
    clf = CalibratedClassifierCV(base, cv=2)
    clf.fit(X_tr_std, y_tr)

    # validation log loss
    proba_val = clf.predict_proba(X_val_std)[:, 1].clip(*PRED_CLIP)
    val_loss = log_loss(y_val, proba_val)
    arm_val_losses.append(val_loss)
    arm_val_ns.append(len(y_val))

    models[arm] = ("sgd", clf)
    scalers[arm] = sc

print("[MODEL] done.")

# supervised 모델 전체 평균 loss (샘플 수로 가중평균)
if arm_val_losses:
    total_samples = np.sum(arm_val_ns)
    weighted_loss = np.sum(
        [l * n for l, n in zip(arm_val_losses, arm_val_ns)]
    ) / total_samples
    print(f"[METRIC] Supervised model average log-loss = {weighted_loss:.4f}")
else:
    print("[METRIC] Supervised model log-loss: N/A (모든 팔이 const 모델)")


# =====================================================
# 2. PROBABILITY MATRIX FOR ENV (P(click | context, arm))
# =====================================================
# env_df 전체 행에 대해 각 arm의 CTR확률을 계산하여 (N_env, K) 행렬 생성
env_df = env_df.sort_index().reset_index(drop=True)
N_env = len(env_df)

num_cols = ["seq_last", "seq_len", "seq_entropy"]
cat_cols = ["gender", "age_group", "day_of_week", "hour"]

X_env_num = env_df[num_cols].values.astype(float)
X_env_cat = env_df[cat_cols].values.astype(float)
X_env_all = np.hstack([X_env_num, X_env_cat])

prob_matrix = np.zeros((N_env, K), dtype=np.float32)

for k, arm in enumerate(arms):
    kind, model = models[arm]
    if kind == "const":
        prob_matrix[:, k] = float(model)
    else:
        sc = scalers[arm]
        clf = model
        X_env_std = sc.transform(X_env_all)
        prob_matrix[:, k] = clf.predict_proba(X_env_std)[:, 1].clip(*PRED_CLIP)

print(f"[ENV] env rows={N_env:,}, prob_matrix shape={prob_matrix.shape}")


# ---------------- CONTEXT FEATURES (FeatureHasher) ----------------
def row_to_dict_fast(r):
    # r: 배열 [gender, age_group, inventory_id, day_of_week, hour, seq_last, seq_len, seq_entropy]
    return {
        "g": f"g_{int(r[0])}",
        "ag": f"ag_{int(r[1])}",
        "inv": f"inv_{int(r[2])}",
        "dow": f"dow_{int(r[3])}",
        "h": f"h_{int(r[4])}",
        "s_last": float(r[5]),
        "s_len": float(r[6]),
        "s_ent": float(r[7]),
    }


ctx_raw = env_df[CTX_COLS].values
ctx_dicts = [row_to_dict_fast(r) for r in ctx_raw]

hasher = FeatureHasher(n_features=HASH_DIM, input_type="dict")
X_ctx = hasher.transform(ctx_dicts).toarray().astype(np.float32)

# scale & normalize
sc_ctx = StandardScaler().fit(X_ctx)
X_std = sc_ctx.transform(X_ctx).astype(np.float32)
X_norm = Normalizer().fit_transform(X_std).astype(np.float32)

d = X_norm.shape[1]
print(f"[ENV] feature dim d={d}")


# =====================================================
# 3. LINUCB AGENT (No Oracle line, baseline = best single arm)
# =====================================================
class LinUCB:
    def __init__(self, K, d, alpha=1.0):
        self.K = K
        self.d = d
        self.alpha = alpha
        self.A_inv = np.stack([np.eye(d, dtype=np.float64) for _ in range(K)], axis=0)
        self.b = np.zeros((K, d), dtype=np.float64)

    def select_arm(self, x):
        """
        x: (d,) numpy vector
        """
        # theta_k = A_k^{-1} b_k
        theta = np.einsum("kij,kj->ki", self.A_inv, self.b)  # (K, d)
        mu = theta @ x  # (K,)
        Ax = self.A_inv @ x  # (K, d)
        s = np.sum(Ax * x, axis=1)  # x^T A^{-1} x
        ucb = mu + self.alpha * np.sqrt(np.maximum(s, 1e-12))
        return int(np.argmax(ucb))

    def update(self, a, x, r):
        """
        a: 선택한 arm index
        x: (d,)
        r: 보상 (0/1)
        """
        A_inv_a = self.A_inv[a]
        Ax = A_inv_a @ x
        denom = 1.0 + float(x @ Ax)
        self.A_inv[a] = A_inv_a - np.outer(Ax, Ax) / denom
        self.b[a] += r * x


# =====================================================
# 4. TRAINING LOOP (PURE LINUCB)
# =====================================================
agent = LinUCB(K=K, d=d, alpha=ALPHA)

counts = np.zeros(K, dtype=np.int64)
sum_rewards = np.zeros(K, dtype=np.float64)
Q = np.zeros(K, dtype=np.float64)

cum_reward = 0.0
ctr_curve = []
regret_curve = []
steps_log = []
counts_snapshots = []

# baseline: best single arm (context 무시)
arm_mean_ctr = prob_matrix.mean(axis=0)  # (K,)
best_arm_idx = int(np.argmax(arm_mean_ctr))
p_star = float(arm_mean_ctr[best_arm_idx])
print(f"[BASELINE] Best single arm = {arms[best_arm_idx]}, CTR={p_star:.4f}")

# policy accuracy용 "oracle best arm" (문맥별 최적 팔, 내부평가용)
oracle_best_arms = np.argmax(prob_matrix, axis=1)

# context-conditioned CTR 용 segment col
ctx_seg_values = env_df[CONTEXT_SEG_COL].values
seg_unique = np.unique(ctx_seg_values)
seg_reward = {v: 0.0 for v in seg_unique}
seg_count = {v: 0 for v in seg_unique}

# policy accuracy count
correct_actions = 0

rng = range(N_env)
if TQDM_AVAILABLE:
    rng = tqdm(rng, desc="Training LinUCB")

for t in rng:
    x = X_norm[t]
    probs = prob_matrix[t]

    # 선택
    arm_idx = agent.select_arm(x)

    # reward 샘플링 (환경은 prob_matrix를 진짜 CTR로 가정)
    r = 1 if np.random.rand() < probs[arm_idx] else 0

    # 업데이트
    agent.update(arm_idx, x, r)

    counts[arm_idx] += 1
    sum_rewards[arm_idx] += r
    Q[arm_idx] = sum_rewards[arm_idx] / max(counts[arm_idx], 1)

    cum_reward += r

    # context-conditioned CTR 기록
    seg_val = ctx_seg_values[t]
    seg_count[seg_val] += 1
    seg_reward[seg_val] += r

    # policy accuracy (oracle best arm 기준)
    if arm_idx == oracle_best_arms[t]:
        correct_actions += 1

    # 로그 포인트 저장
    step = t + 1
    if step % LOG_EVERY == 0 or step == 1:
        ctr = cum_reward / step
        regret = step * p_star - cum_reward
        ctr_curve.append(ctr)
        regret_curve.append(regret)
        steps_log.append(step)
        counts_snapshots.append(counts.copy())

        print(f"[STEP {step:6d}] CTR={ctr:.4f}, Regret={regret:.1f}")

print("\n[RESULT] LinUCB Final CTR = {:.4f}".format(cum_reward / N_env))

# 최종 regret
final_regret = N_env * p_star - cum_reward
print(f"[RESULT] Final cumulative regret vs best single arm = {final_regret:.1f}")
print(f"[RESULT] Average regret per step = {final_regret / N_env:.4f}")

print("\n[Arm Q (final) vs static mean CTR]")
for k, arm in enumerate(arms):
    print(
        f"  arm={arm:2d}  count={counts[k]:7d}, "
        f"Q={Q[k]:.4f},  static_mean={arm_mean_ctr[k]:.4f}"
    )


# =====================================================
# 5. EVALUATION METRICS
# =====================================================

# (1) Cumulative CTR
plt.figure(figsize=(6, 4))
plt.plot(steps_log, ctr_curve, label="LinUCB")
plt.title("Cumulative CTR — Contextual LinUCB")
plt.xlabel("Step")
plt.ylabel("Cumulative CTR")
plt.grid(True, ls="--", alpha=0.5)
plt.tight_layout()
plt.show()

# (2) Cumulative Regret vs best single arm
plt.figure(figsize=(6, 4))
plt.plot(steps_log, regret_curve)
plt.title("Cumulative Regret — LinUCB (vs best single arm)")
plt.xlabel("Step")
plt.ylabel("Regret")
plt.grid(True, ls="--", alpha=0.5)
plt.tight_layout()
plt.show()

# (3) Estimated Q-value convergence: True CTR vs Q(a)
x_idx = np.arange(K)
plt.figure(figsize=(8, 4))
bar_w = 0.35
plt.bar(x_idx - bar_w / 2, arm_mean_ctr, width=bar_w, label="True CTR (static mean)")
plt.bar(x_idx + bar_w / 2, Q, width=bar_w, label="LinUCB Q(a)")
plt.xticks(x_idx, [str(a) for a in arms])
plt.xlabel("arm (ad_contents)")
plt.ylabel("CTR")
plt.title("Per-arm CTR — True vs LinUCB Estimated Q")
plt.grid(True, axis="y", ls="--", alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()

# (4) Action Selection Distribution — counts
plt.figure(figsize=(8, 4))
plt.bar(x_idx, counts)
plt.xticks(x_idx, [str(a) for a in arms])
plt.xlabel("arm (ad_contents)")
plt.ylabel("Selection count")
plt.title("Final Arm Selection Counts — LinUCB")
plt.grid(True, axis="y", ls="--", alpha=0.5)
plt.tight_layout()
plt.show()

# (4-2) 시간에 따른 누적 선택 비율
if counts_snapshots:
    counts_snapshots = np.vstack(counts_snapshots)   # (L, K)
    steps_arr = np.array(steps_log)
    ratios = counts_snapshots / steps_arr[:, None]   # (L, K)

    plt.figure(figsize=(9, 5))
    for k, arm in enumerate(arms):
        plt.plot(steps_arr, ratios[:, k], label=f"arm {arm}")
    plt.xlabel("Step")
    plt.ylabel("Selection ratio (cumulative)")
    plt.title("Arm Selection Ratio over Time — LinUCB")
    plt.grid(True, ls="--", alpha=0.5)
    plt.legend()
    plt.tight_layout()
    plt.show()

# (5) Context-conditioned CTR (예: day_of_week 기준)
print(f"\n[METRIC] Context-conditioned CTR by '{CONTEXT_SEG_COL}'")
rows = []
for v in sorted(seg_unique):
    c = seg_count[v]
    r = seg_reward[v]
    ctr_seg = r / c if c > 0 else 0.0
    rows.append((v, c, r, ctr_seg))
    print(f"  {CONTEXT_SEG_COL}={v:>2}  n={c:6d}, clicks={int(r):5d}, CTR={ctr_seg:.4f}")

# 원하면 그래프로도 표현 (선택)
plt.figure(figsize=(7, 4))
plt.bar(
    [str(v) for v in sorted(seg_unique)],
    [row[3] for row in rows],
)
plt.xlabel(CONTEXT_SEG_COL)
plt.ylabel("CTR")
plt.title(f"Context-conditioned CTR by {CONTEXT_SEG_COL} — LinUCB")
plt.grid(True, axis="y", ls="--", alpha=0.5)
plt.tight_layout()
plt.show()

# (6) Policy Accuracy (oracle best arm 기준)
policy_accuracy = correct_actions / N_env
print(f"\n[METRIC] Policy Accuracy (vs context-wise best arm) = {policy_accuracy:.4f}")

# (7) Supervised model average log-loss는 앞에서 이미 출력함
```

```python
# -*- coding: utf-8 -*-
"""
Contextual Bandit — LinUCB with multi-seed evaluation
- Toss_preprocessed_FF.csv 기반
- Reward model (per-arm logistic) + LinUCB 정책
- 여러 random seed에 대해 Final CTR을 측정하고
  평균 및 95% 신뢰구간을 계산 + 시각화
"""

import numpy as np
import pandas as pd
from tqdm import tqdm
import matplotlib.pyplot as plt

from sklearn.linear_model import SGDClassifier
from sklearn.calibration import CalibratedClassifierCV
from sklearn.preprocessing import StandardScaler, Normalizer
from sklearn.feature_extraction import FeatureHasher

# ===================== 공통 설정 =====================
PATH      = "Toss_preprocessed_FF.csv"

CTX_COLS  = [
    "gender", "age_group", "inventory_id",
    "day_of_week", "hour",
    "seq_last", "seq_len", "seq_entropy"
]

ARM_COL    = "ad_contents"
REWARD_COL = "reward"

TRAIN_RATIO   = 0.7          # model_df : env_df = 70 : 30
HASH_DIM_CTX  = 128          # Context FeatureHasher 차원
PRED_CLIP     = (1e-3, 1-1e-3)

# Reward model (SGDClassifier) 설정
SGD_ALPHA     = 1e-4
SGD_MAX_ITER  = 200
SGD_TOL       = 1e-3

# LinUCB 설정
ALPHA      = 1.0             # 탐색 계수
LOG_EVERY  = 20000           # 로그 출력 스텝 간격


# =====================================================
# 1. 한 번의 LinUCB 실험을 수행하는 함수
#    - seed를 인자로 받아, Final CTR(float)를 반환
# =====================================================
def run_linucb_once(seed: int = 42,
                    verbose: bool = False,
                    log_every: int = LOG_EVERY) -> float:
    """
    주어진 random seed로 LinUCB를 1회 실행하고,
    최종 CTR (float)을 반환.
    verbose=True 이면 세부 로그/지표도 출력.
    """
    rng = np.random.default_rng(seed)

    # ---------- (1) 데이터 로드 및 전처리 ----------
    df = pd.read_csv(PATH)
    df = df.dropna(subset=CTX_COLS + [ARM_COL, REWARD_COL]).reset_index(drop=True)
    df[REWARD_COL] = df[REWARD_COL].astype(int)

    n_total = len(df)
    if verbose:
        print(f"[INFO] total rows={n_total:,}")

    # 재현성을 위해 셔플
    df = df.sample(frac=1.0, random_state=seed).reset_index(drop=True)

    # train(model) / env(split)
    split_idx = int(n_total * TRAIN_RATIO)
    model_df = df.iloc[:split_idx].reset_index(drop=True)
    env_df   = df.iloc[split_idx:].reset_index(drop=True)

    if verbose:
        print(f"[SPLIT] model rows={len(model_df):,}, env rows={len(env_df):,}")

    # ---------- (2) 팔 정보 ----------
    arms = np.sort(model_df[ARM_COL].unique())
    K = len(arms)

    # ---------- (3) 팔별 Reward 모델 학습 ----------
    models  = {}
    scalers = {}
    total_logloss = 0.0
    total_count   = 0

    for arm in arms:
        sub = model_df[model_df[ARM_COL] == arm]
        y   = sub[REWARD_COL].values
        n_a = len(sub)
        mean_ctr = float(y.mean()) if n_a > 0 else 0.0

        if verbose:
            print(f"[MODEL] arm={arm} n={n_a:,}, mean_ctr={mean_ctr:.4f}")

        if n_a < 200:
            # 데이터가 너무 적으면 상수 확률 모델
            p_const = float(np.clip(mean_ctr, *PRED_CLIP))
            models[arm] = ("const", p_const)
            # 상수 모델의 log-loss
            eps = 1e-15
            p  = np.clip(p_const, eps, 1-eps)
            ll = -(y * np.log(p) + (1-y) * np.log(1-p)).sum()
            total_logloss += ll
            total_count   += n_a
            continue

        # numeric / categorical 분리
        num = sub[["seq_last", "seq_len", "seq_entropy"]].values
        sc  = StandardScaler().fit(num)
        num_scaled = sc.transform(num)

        cat = sub[["gender", "age_group", "day_of_week", "hour"]].values
        X_small = np.hstack([num_scaled, cat])

        base = SGDClassifier(
            loss="log_loss",
            alpha=SGD_ALPHA,
            max_iter=SGD_MAX_ITER,
            tol=SGD_TOL,
            random_state=seed,
        )
        clf = CalibratedClassifierCV(base, cv=2)
        clf.fit(X_small, y)

        models[arm]  = ("sgd", clf)
        scalers[arm] = sc

        # log-loss 계산
        prob = clf.predict_proba(X_small)[:, 1]
        prob = np.clip(prob, 1e-15, 1 - 1e-15)
        ll   = -(y * np.log(prob) + (1-y) * np.log(1-prob)).sum()
        total_logloss += ll
        total_count   += n_a

    if total_count > 0 and verbose:
        avg_logloss = total_logloss / total_count
        print(f"[METRIC] Supervised model average log-loss = {avg_logloss:.4f}")

    # ---------- (4) env_df에 대한 클릭확률 prob_matrix 계산 ----------
    env_num = env_df[["seq_last", "seq_len", "seq_entropy"]].values
    env_cat = env_df[["gender", "age_group", "day_of_week", "hour"]].values
    n_env   = len(env_df)

    prob_matrix = np.zeros((n_env, K), dtype=np.float32)

    for k, arm in enumerate(arms):
        kind = models[arm][0]
        if kind == "const":
            prob_matrix[:, k] = models[arm][1]
        else:
            sc  = scalers[arm]
            clf = models[arm][1]
            num_scaled = sc.transform(env_num)
            X_small    = np.hstack([num_scaled, env_cat])
            prob_matrix[:, k] = clf.predict_proba(X_small)[:, 1].clip(*PRED_CLIP)

    if verbose:
        print(f"[ENV] env rows={n_env:,}, prob_matrix shape={prob_matrix.shape}")

    # ---------- (5) 컨텍스트 피처 해싱 및 정규화 ----------
    def row_to_dict_fast(r):
        return {
            "g":   f"g_{int(r[0])}",
            "ag":  f"ag_{int(r[1])}",
            "inv": f"inv_{int(r[2])}",
            "dow": f"dow_{int(r[3])}",
            "h":   f"h_{int(r[4])}",
            "s_last": float(r[5]),
            "s_len":  float(r[6]),
            "s_ent":  float(r[7]),
        }

    ctx_raw   = env_df[CTX_COLS].values
    ctx_dicts = [row_to_dict_fast(r) for r in ctx_raw]

    hasher_ctx = FeatureHasher(n_features=HASH_DIM_CTX, input_type="dict")
    X_ctx      = hasher_ctx.transform(ctx_dicts).toarray().astype(np.float32)

    scaler_ctx = StandardScaler()
    X_std  = scaler_ctx.fit_transform(X_ctx).astype(np.float32)
    X_norm = Normalizer().fit_transform(X_std).astype(np.float32)

    d = X_norm.shape[1]
    if verbose:
        print(f"[ENV] feature dim d={d}")

    # ---------- (6) Baseline: best single arm ----------
    static_ctr_by_arm = prob_matrix.mean(axis=0)
    best_arm = int(static_ctr_by_arm.argmax())
    best_ctr = float(static_ctr_by_arm[best_arm])

    if verbose:
        print(f"[BASELINE] Best single arm = {best_arm}, CTR={best_ctr:.4f}")

    # ---------- (7) LinUCB 초기화 ----------
    A_inv = np.array([np.eye(d, dtype=np.float32) for _ in range(K)])  # (K, d, d)
    b     = np.zeros((K, d), dtype=np.float32)

    counts      = np.zeros(K, dtype=np.int64)
    sum_rewards = np.zeros(K, dtype=np.float64)

    cum_reward  = 0.0
    ctr_curve   = []
    regret_curve= []

    # policy accuracy 측정용
    correct_policy = 0

    # ---------- (8) 메인 루프 ----------
    if verbose:
        print(f"[RUN] start LinUCB (seed={seed})")

    for t in tqdm(range(n_env), desc=f"Training LinUCB (seed={seed})"):
        x = X_norm[t]                # (d,)
        probs_t = prob_matrix[t]     # (K,)

        # LinUCB 선택 규칙
        theta = A_inv @ b[..., None]      # (K, d, 1)
        theta = theta.squeeze(2)          # (K, d)
        mu    = theta @ x                 # (K,)
        Ax    = A_inv @ x                 # (K, d)
        bonus = ALPHA * np.sqrt((Ax * x).sum(axis=1))   # (K,)
        score = mu + bonus
        a     = int(score.argmax())

        # Environment에서 reward 샘플링
        r = 1 if rng.random() < probs_t[a] else 0

        # LinUCB 업데이트 (Sherman–Morrison)
        A = A_inv[a]
        Ax_a = A @ x
        denom = 1.0 + float(x @ Ax_a)
        A_inv[a] = A - np.outer(Ax_a, Ax_a) / denom
        b[a]    += r * x

        # 통계 업데이트
        counts[a]      += 1
        sum_rewards[a] += r
        cum_reward     += r

        # CTR / Regret
        step_ctr    = cum_reward / (t + 1)
        step_regret = (t + 1) * best_ctr - cum_reward

        ctr_curve.append(step_ctr)
        regret_curve.append(step_regret)

        # policy accuracy: reward 모델 기준 best arm vs LinUCB 선택 arm
        best_arm_ctx = int(probs_t.argmax())
        if best_arm_ctx == a:
            correct_policy += 1

        if verbose and (t + 1) % log_every == 0:
            print(f"[STEP {t+1:6d}] CTR={step_ctr:.4f}, Regret={step_regret:.1f}")

    final_ctr     = cum_reward / n_env
    final_regret  = regret_curve[-1]
    avg_regret    = final_regret / n_env
    policy_acc    = correct_policy / n_env

    if verbose:
        print(f"\n[RESULT] LinUCB Final CTR = {final_ctr:.4f}")
        print(f"[RESULT] Final cumulative regret vs best single arm = {final_regret:.1f}")
        print(f"[RESULT] Average regret per step = {avg_regret:.4f}")
        print(f"[RESULT] Policy Accuracy (vs context-wise best arm) = {policy_acc:.4f}")

        # 팔별 요약 출력 (원하면 유지)
        print("\n[Arm Q (final) vs static mean CTR]")
        Q_est = np.zeros(K, dtype=np.float64)
        for k in range(K):
            Q_est[k] = sum_rewards[k] / counts[k] if counts[k] > 0 else 0.0
            print(
                f"  arm={arms[k]:2}  count={counts[k]:7d}, "
                f"Q={Q_est[k]:.4f},  static_mean={static_ctr_by_arm[k]:.4f}"
            )

        # (선택) 단일 시드용 CTR / Regret 곡선 시각화
        plt.figure(figsize=(6,4))
        plt.plot(ctr_curve)
        plt.title("Cumulative CTR — LinUCB")
        plt.xlabel("Step")
        plt.ylabel("CTR")
        plt.grid(True, ls="--", alpha=0.5)
        plt.tight_layout()
        plt.show()

        plt.figure(figsize=(6,4))
        plt.plot(regret_curve)
        plt.title("Cumulative Regret vs Best Single Arm — LinUCB")
        plt.xlabel("Step")
        plt.ylabel("Regret")
        plt.grid(True, ls="--", alpha=0.5)
        plt.tight_layout()
        plt.show()

    return final_ctr


# =====================================================
# 2. 여러 seed에 대해 LinUCB를 반복 실행하고
#    평균 CTR 및 95% 신뢰구간 계산 + 시각화
# =====================================================
if __name__ == "__main__":
    SEEDS = list(range(10))       # 예: 0 ~ 9까지 10회 반복
    ctrs  = []

    for s in SEEDS:
        print(f"\n[RUN] seed={s}")
        # verbose=False 로 두면 로그 최소화. 필요하면 True로 한 번만 자세히 봐도 됨.
        ctr = run_linucb_once(seed=s, verbose=False)
        ctrs.append(ctr)
        print(f"[RUN RESULT] seed={s}, Final CTR={ctr:.4f}")

    ctrs = np.array(ctrs)
    mean_ctr = ctrs.mean()
    std_ctr  = ctrs.std(ddof=1)
    m        = len(ctrs)
    ci95     = 1.96 * std_ctr / np.sqrt(m)
    ci_low, ci_high = mean_ctr - ci95, mean_ctr + ci95

    print("\n========== LinUCB Multi-seed Evaluation ==========")
    print(f"Seeds         : {SEEDS}")
    print(f"Mean CTR      : {mean_ctr:.4f}")
    print(f"Std of CTR    : {std_ctr:.4f}")
    print(f"95% CI (CTR)  : [{ci_low:.4f}, {ci_high:.4f}]")
    print("--------------------------------------------------")
    for s, c in zip(SEEDS, ctrs):
        print(f"  seed={s:2d} -> CTR={c:.4f}")

    # ================= 시각화 추가 부분 =================

    # (1) seed별 CTR 막대그래프 + 평균 / 95% CI
    plt.figure(figsize=(7,4))
    x = np.arange(m)
    plt.bar(x, ctrs)
    plt.xticks(x, SEEDS)
    plt.xlabel("Random seed")
    plt.ylabel("Final CTR")
    plt.title("LinUCB Final CTR by Seed")

    # 평균 CTR 수평선
    plt.axhline(mean_ctr, linestyle="--", linewidth=1.5, label=f"Mean CTR = {mean_ctr:.4f}")
    # 95% 신뢰구간 상·하한
    plt.axhline(ci_low, linestyle=":", linewidth=1.2, label=f"95% CI low  = {ci_low:.4f}")
    plt.axhline(ci_high, linestyle=":", linewidth=1.2, label=f"95% CI high = {ci_high:.4f}")

    plt.legend()
    plt.grid(True, ls="--", alpha=0.5)
    plt.tight_layout()
    plt.show()

    # (2) CTR 분포 박스플롯 (변동성 한눈에 보기)
    plt.figure(figsize=(4,4))
    plt.boxplot(ctrs, vert=True, labels=["LinUCB"])
    plt.ylabel("Final CTR")
    plt.title("Distribution of LinUCB Final CTR across seeds")
    plt.grid(True, axis="y", ls="--", alpha=0.5)
    plt.tight_layout()
    plt.show()
```
